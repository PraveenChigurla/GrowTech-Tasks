**### 1. Create a list with numbers 1 to 10 in it ([1,2,3….10]) Write a for loop on this list to sum up all numbers in the list except 5 and 7. Print the result**

```python
numb=[i for i in range(1,11)]
print(numb)
sum=0
for i in numb:
  if i==5 or i==7:
    continue
  else:
    sum+=i
print(sum)
```

### 2. Write a while loop to create a list with numbers from 1 to 100. When inserting new number in the list, skip every number which is divisible by 3. Print the result. Tip: To check whether number is divisible by 3 you can use % operator

```python
i=1
print(i)
new_list=[]
while i<=100:
  if i%3==0:
    i+=1
    continue
  new_list.append(i)
  i+=1
print(new_list)
```

### Check if a given number is prime

```python
x=int(input("enter the number: "))
fact=0
for i in range(1,x+1):
  if x%i==0:
    fact+=1
if fact==2:
  print("Prime")
else:
  print("Not a Prime")
```

### Find the factorial of a number

```python
def fact(n):
  if n==1:
    return 1
  else:
    return n* fact(n-1)

n=int(input("enter the number: "))
print(fact(n))
```

### Generate the Fibonacci sequence up to a certain number of terms

```python
fibos=[]
def fibo(n):
  if n==0:
    return 0
  elif n==1:
    return 1
  else:
    return fibo(n-1) + fibo(n-2)

n=int(input("which positioned number in the fibonacci series you want to know?: "))

for i in range(n):
  fibos.append(fibo(i))

print(fibos[n-1])
```

### Check if a word is a palindrome

```python
string= input("enter the word: ")
rev=''

for i in string[-1::-1]:
  rev+=i
print(rev)

if rev==string:
  print("Palindrome")
print("Not a Palindrome")

```

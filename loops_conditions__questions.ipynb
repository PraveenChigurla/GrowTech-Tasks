```python
# Question 3
3. Write a program to check if a given number is prime.
```

```python
def prime(x):
  count=0
  for i in range(1,x+1):
    if (x%i)==0:
      count+=1
  if count>2:
    return "Not a Prime number "
  else:
    return "Prime Number"

x=int(input("enter a number: "))
prime(x)
```

```python
# Question 4
4. Print the Fibonacci sequence up to n terms using a loop.
```

```python
x= int(input("enter a number: "))
fibo=[]
for i in range(x):
  if i==0:
    #y=0
    fibo.append(0)
  elif i==1:
    #y=1
    fibo.append(1)
  else:
    #y=fibo[i-1]+ fibo[i-2]
    fibo.append(fibo[i-1]+ fibo[i-2] )
print(fibo)
print(fibo[x-1])
```

```python
#recursion for printing n'th number in a fibonacci series
def fibonacci(x):
  if x==0:
    return 0
  elif x==1:
    return 1
  else:
    return fibonacci(x-1)+ fibonacci(x-2)

x= int(input("enter a number: "))
y= fibonacci(x-1)
print(y)
```

```python
# Question 8
8. Use a loop to reverse an integer (e.g., 123 → 321).
```

```python
x= int(input("enter a number: "))
rev=0

while True:
  if x<10:
    rev=rev*10+x
    break
  else:
    y=x%10
    rev=rev*10+y
    x=int(x/10)

print(rev)
```

```python
# Question 9
9. Write a program that uses a loop to check if a string is a palindrome.
```

```python
x=input("Enter a string: ")
rev=''

for i in range(len(x)-1,-1,-1):
    #print(x[i])
    rev+=x[i]
print(rev)

if x==rev:
  print("Palindrome")
else:
  print("Not Palindrome")


```

```python
# Question 12
12. Print a right-angled triangle pattern using `*` for n rows.
```

```python
#Filled Right-Angled Triangle
x=int(input("enter no. of rows: "))
for i in range(1,x+1):
    for j in range(i):
        print("*", end=" ")
    print()
```

```python
#Hallow Right-Angled Triangle
x=int(input("enter no. of rows: "))
for i in range(1,x+1):
    for j in range(i):
      if j==0 or j==(i-1) or i==x:
        print("*", end=" ")
      else:
        print(" ", end=" ")
    print()
```

```python
n = 4
cols = 20

# create empty matrix
mat = [[" " for _ in range(cols)] for _ in range(n)]

row = 0
down = True

for col in range(cols):

    mat[row][col] = "*"

    # change direction
    if row == n - 1:
        down = False

    elif row == 0:
        down = True

    # move row
    if down:
        row += 1
    else:
        row -= 1

# print matrix
for r in mat:
    print(" ".join(r))
```

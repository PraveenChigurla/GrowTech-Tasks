```python
import pandas as pd
import requests
import json
from pandas import json_normalize
```

```python
response=requests.get("https://www.risingsigma.com/zippy-v2/json.php?shop=497b2f-97", allow_redirects=True)
print(response.headers.get('content-type'))
content=response.content
content
```

```python
type(response)
```

```python
data = json.loads(content) ## parse a JSON string and convert it into a Python Dictionary.
data
```

```python
type(data)
```

```python
data = json_normalize(data)# convert dictionary to dataframe
data
```

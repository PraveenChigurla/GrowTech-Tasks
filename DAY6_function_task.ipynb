11. Write a function that returns the nth Fibonacci number.

```python
x=5 #n'th number
fibo=lambda n:0 if n==0 else 1 if n==1 else fibo(n-1)+fibo(n-2)

print(list(map(fibo, range(x)))[x-1])
```

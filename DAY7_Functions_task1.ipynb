24. Create a function `binary_search` that implements binary search on a sorted list.

```python
l = [1, 2, 3, 4, 6, 7, 8, 9, 110, 302, 2324]

target = 8

start = 0
end = len(l) - 1

while start <= end:
    mid = (start + end) // 2

    if l[mid] == target:
        print("Element found at index:", mid)
        break

    elif target < l[mid]:
        end = mid - 1

    else:
        start = mid + 1
```

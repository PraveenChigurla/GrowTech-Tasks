27.	What happens if you try to access a key that doesn’t exist? How can you avoid an error?

```python
# If you try to access a key that does not exist using standard square brackets (dict[key]), Python raises a KeyError and stops your program from running.

user = {"name": "Alice"}
# print(user["age"])  # ❌ Raises KeyError: 'age'

# The .get() method returns None instead of throwing an error if the key is missing. You can also specify a custom default fallback value as the second argument.pythonuser = {"name": "Alice"}

# Returns None safely
age = user.get("age")
print(age)
# Returns 25 as a fallback default
age_with_default = user.get("age", 25)
print(age_with_default)
```

32.	Write a program to merge two dictionaries.

```python
# Method 1: Using the Merge Operator '|'
# Note: Second dict overwrites first

dict1 = {'a': 1, 'b': 2}
dict2 = {'b': 99, 'c': 4}

# Merges both into a brand new dictionary
merged_dict = dict1 | dict2

print(merged_dict)
# Output: {'a': 1, 'b': 99, 'c': 4} (Note: 'b' was updated to 99)
```

```python
#Method 2: using "update()"" If you want to merge the second dictionary directly into the first one without creating a new variable, use .update().
# Note: Second dict overwrites first

dict1 = {'a': 1, 'b': 2}
dict2 = {'b': 99, 'c': 4}

# Modifies dict1 directly
dict1.update(dict2)

print(dict1)
# Output: {'a': 1, 'b': 99, 'c': 4}
```

34.	Write a Python program to sort a dictionary by its values.

```python
st={'Hari': 22, 'Mani': 23, 'Praveen': 14, 'sid': 20, 'uday': 19, 'Harish': 25, 'Manish': 23}

x=[(k,v) for k,v in st.items()]
print(x)

for k in range(len(x)):
  for i in range(0,len(x)-1):
    if (x[i][1]) > (x[i+1][1]):
      x[i+1],x[i]=x[i],x[i+1]
print(x)
y=dict(x)
print(y)
```

```python
st={'Hari': 22, 'Mani': 23, 'Praveen': 14, 'sid': 20, 'uday': 19, 'Harish': 25, 'Manish': 23}

# 1. Sort by values (x[1] represents the value in the key-value tuple)
sorted_pairs = sorted(st.items(), key=lambda x: x[1])
print(sorted_pairs)

# 2. Convert the list of sorted tuples back into a dictionary
sorted_dict = dict(sorted_pairs)

print(sorted_dict)
```

```python
scores = {'Alice': 85, 'Bob': 92, 'Charlie': 78, 'Diana': 95}

# Sort by values in descending order
sorted_descending = dict(sorted(scores.items(), key=lambda x: x[1], reverse=True))

print(sorted_descending)
# Output: {'Diana': 95, 'Bob': 92, 'Alice': 85, 'Charlie': 78}
```

```python
# Method 3: Using a Dictionary Comprehension

scores = {'Alice': 85, 'Bob': 92, 'Charlie': 78, 'Diana': 95}

sorted_dict = {k: v for k, v in sorted(scores.items(), key=lambda val: val[1])}
print(sorted_dict)
```

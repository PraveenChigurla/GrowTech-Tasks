Slicing:
Slicing involves extracting a contiguous subset or a range of elements from a sequence or data structure.

Basic Slicing: This is performed using the [start:stop:step] syntax.

start: The index where the slice begins (inclusive). If omitted, it defaults to the beginning of the sequence.

stop: The index where the slice ends (exclusive). If omitted, it defaults to the end of the sequence.

step: The increment between elements in the slice. If omitted, it defaults to 1.

```python
slicing---[start_num:stop_num:step]
```

# 2. Tuple:
- Tuples are similar to lists, but they are immutable, meaning once you
- create a tuple, you cannot change its elements.
- Can contain items of different types.
- They are created using  parenthesis.
- Example: (1, 'a', 3.14, 'c', 'E')
- Tuples are useful when you want to ensure the data cannot be altered.

```python
#items cannot be removed from tuple
#conversion to list is workaround
data_tuple = ("data", "science", "machine", "learning")
data_tuple = list(data_tuple)
data_tuple.remove("science")
data_tuple = tuple(data_tuple)
data_tuple
```

# 3. Set:
- Sets are unordered collections of unique elements.
- They are created using curly braces.
- Example: {'a', 'b', 1, 2, 3, 'c'}
- Sets are useful for storing elements without duplicates.  
- Sets are great for handling unique items and mathematical set operations.


```python
data_set={1,5,8,8,5,7}
type(data_set)
```

```python
data_set
```

```python
data_set = {"data", "science", "machine", "learning"}
print(data_set)
print(type(data_set))
```

```python
a
```

```python
#sets do not hold duplicate values

data_set = {"data", "science", "machine", "learning", "science"}
print(data_set)
```

```python
print(data_set[0]) # sets are unordered, cannot be accessed by index
```

```python
#iterating over set
data_set = {"data", "science", "machine", "learning"}

for value in data_set:
  print(value)
```

#### add(): adds a single element to the set.

```python

data_set = {"data", "science", "machine", "learning"}

data_set.add("maths")

print(data_set)
```

#### update(): adds elements from another set to the existing set

```python
# Add two sets
data_set = {1,1.22,"data", "science", "machine", "learning"}
machine_set = {"lr", "svm", "kmeans"}

data_set.update(machine_set)

print(data_set)
```

#### remove(): removes a specified element (raises an error if not found).

```python

data_set = {"data", "science", "machine", "learning"}

data_set.remove("machine")

print(data_set)
```

#### pop(): removes random item from set as its unordered

```python
data_set = {"data", "science", "machine", "learning"}

data_set.pop()

print(data_set)
```

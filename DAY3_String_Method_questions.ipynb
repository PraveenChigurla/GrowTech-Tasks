```python
print('''✅ Returns True for:
Standard digits: "0", "1", "99"
Fractions: "½", "¾", "¼"
Superscripts & Subscripts: "²", "³", "₁"
Roman numerals: "Ⅰ", "Ⅳ", "Ⅹ" (must be Unicode Roman numeral characters, not standard letters like "X" or "V")
Numbers in other languages: "一", "二", "三" (Chinese/Japanese numerals)

❌ Returns False for:
Decimals / Floats: "3.14", "0.5" (the period . is not a numeric character)
Negative numbers: "-5", "-10" (the hyphen - is not a numeric character)
Alphabetic letters: "abc", "xyz"
Alphanumeric text: "123a", "id-5"
Empty strings: ""Whitespace: "   ", "\\n", "\\t" ''')
```

Q20. How does `.casefold()` differ from `.lower()`?

```python
# lower() converts all the alphabets into lowercase and
# casefold can do that but also it can bringdown the international language letters to respective english alphabetic form

text = "Straße"  # German word for Street

print(text.lower())     # "straße"  (Keeps the 'ß')
print(text.casefold())  # "strasse" (Converts 'ß' to 'ss')

```


Q22. Demonstrate the use of f-strings and compare them with `.format()`.

```python
# F-strings allow execution of code directly within the placeholder. .format() cannot do this.

text = "python"
# F-string handles logic inline
print(f"Uppercase: {text.upper()}")

# .format() requires passing the executed result
print("Uppercase: {}".format(text.upper()))
```

```python
# F-strings are roughly 2x to 3x faster because they are parsed by the compiler into efficient bytecodes rather than running a string-parsing function.

name="Praveen"
age=22
print("Hello {}, you are {}".format(name,age))
print(f"Hello {name}, you are {age}")
```

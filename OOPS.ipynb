Create a class BankAccount.

Requirements:

account_holder
balance

Methods:

*deposit(amount)
*withdraw(amount)
*check_balance()

Conditions:

Cannot withdraw more than balance.

Concepts:

Encapsulation mindset
State modification

```python
class BankAccount:
  def __init__(self,account_holder,balance=0):
    self.name=account_holder
    self.bal=balance
    print(f"Hi {self.name}. Your account balance is: {self.bal}RS ")

  def deposit(self,amt):
    # amt=input
    self.bal+=amt
    return f"Deposited Successfully. Your account balance is: {self.bal}RS "

  def withdraw(self,amt):
    if amt<=self.bal:
      self.bal-=amt
      print(f"Withdraw Successfully. Your account balance is: {self.bal}RS ")
    else:
      print("Insufficient Balance")

  def check_balance(self):
    print(f"Your account balance is: {self.bal}RS ")


C1=BankAccount("Pavan",5000)
C1.check_balance()

C1.withdraw(8000)
C1.check_balance()

C1.deposit(5000)
C1.check_balance()

C1.withdraw(8000)
C1.check_balance()
```

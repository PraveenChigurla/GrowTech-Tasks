
15. How do average house prices vary depending on the number of bedrooms?


```python
import matplotlib.ticker as ticker

plt.figure(figsize=(8,5))
x=df.groupby(['bedrooms'])['price'].mean().round(2)
xticks=x.index
yticks=x.values

plt.xticks(rotation=(30))
plt.xticks(np.arange(0,100,10))
plt.yticks(np.arange(0,10000000,1000000))
plt.bar(xticks,yticks, color=np.random.choice(['magenta','green','red','cyan','brown','gold','blue','black','chocolate'],6,replace=False))

plt.gca().yaxis.set_major_formatter(
    ticker.StrMethodFormatter('{x:,.0f}')
)

plt.bar_label((plt.gca().containers[0]),fmt="%.1f")
plt.show()
```


18. What is the trend of average price with respect to house area?


```python
# Create 10 area bins
df['area_bin'] = pd.cut(df['area'], bins=10)

# Average price in each bin
avg_price = df.groupby('area_bin', observed=False)['price'].mean()

plt.figure(figsize=(10,5))
plt.plot(avg_price.index.astype(str), avg_price.values, marker='o')
plt.xticks(rotation=45)

plt.xlabel('Area Range')
plt.ylabel('Average Price')
plt.title('Trend of Average Price with Respect to House Area')
plt.grid(True)
plt.show()
```

```python
# Create 10 area bins for a more robust trend analysis
df['area_bin_for_plot'] = pd.cut(df['area'], bins=10)

# Calculate the average price for each area bin
avg_price_by_area_bin = df.groupby('area_bin_for_plot', observed=False)['price'].mean().round(2)

plt.figure(figsize=(12, 6))
# Plot as a bar chart to maintain the original intent of the cell
avg_price_by_area_bin.plot(kind='bar', color='skyblue')

plt.xlabel('Area Range')
plt.ylabel('Average Price')
plt.title('Trend of Average Price with Respect to House Area')
plt.xticks(rotation=45, ha='right') # Rotate labels for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout() # Adjust layout to prevent labels from being cut off
plt.show()

# Drop the temporary 'area_bin_for_plot' column
df = df.drop(columns=['area_bin_for_plot'])
```

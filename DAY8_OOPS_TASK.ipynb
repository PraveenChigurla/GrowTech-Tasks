   Create two accounts and demonstrate deposits and withdrawals.

```python
class BankAccount:
  def __init__(self,holder,balance):
    self.name=holder
    self.bal=balance

  def __str__(self):
    return f"\nWelcome {self.name} \n Balance:{self.bal}"

  def deposit(self,amt):
    print(self.__str__())
    self.bal+=amt
    return f"Deposited {amt} Successfully! Current balance={self.bal}"

  def withdraw(self,amt):
    print(self.__str__())
    if amt>self.bal:
      print("Insufficient Balance")
    else:
      self.bal-=amt
      print(f"Withdrawn {amt} Successfully! Current balance={self.bal}")

x=BankAccount("Naruto",50000)
print(x.deposit(10000))
x.withdraw(10000000)
```

```python
class Restaurant:
    def __init__(self):
        self.Rests = {
            'Mehfil': {'location': 'Narayanguda', 'contact_no': '90545 60503'},
            'Pista House': {'location': 'Uppal', 'contact_no': '98875 04654'},
            'Shadab': {'location': 'charminar', 'contact_no': '0127784150013'}
        }


class menuItem(Restaurant):
    def __init__(self):
        super().__init__()

        self.menu = {
            "Mehfil": {
                "items": {"Biryani": 160, "Chicken Mandi": 320}
            },
            "Pista House": {
                "items": {"Haleem": 250, "Biryani": 180}
            },
            "Shadab": {
                "items": {"Qurbani ka Meetha": 150, "Biryani": 200}
            }
        }


class Customer:
    def __init__(self):
        self.cname = input("enter your name: ")
        self.cphoneNo = input("enter your phone number: ")


class order(menuItem, Customer):
    def __init__(self):
        menuItem.__init__(self)
        Customer.__init__(self)

        self.orders = {}
        self.count = 0

    def add(self):

        print("-------------------------Restaurants-----------------------------")

        for i in self.Rests:
            print(i, "===>>", self.Rests[i])

        r = input("\n Choose Restaurant: ")

        if r in self.menu:

            print("\n MENU ITEMS: ")
            print(self.menu[r]["items"])

            item = input("\n Choose Item: ")

            if item in self.menu[r]["items"]:

                q = int(input("\n Enter Quantity : "))

                price = self.menu[r]["items"][item]

                # initialize restaurant cart
                if r not in self.orders:
                    self.orders[r] = []

                # check if item already exists
                found = False

                for i in range(len(self.orders[r])):

                    if self.orders[r][i][0] == item:

                        self.orders[r][i][2] += q
                        found = True
                        break

                if found == False:
                    self.orders[r].append([item, price, q])

                print(f"-------Item added successfully-----")
                print(f'\n Your Bucket={self.orders}')

            else:

                print("-------Item with that name not found in this restaurant! Try other items or try other Restaurants-----")

        else:
            print("-------Restaurant not found!------")

    def remove(self):

        r = input("enter Restaurant name: ")
        item = input("Item Name: ")

        if r in self.orders:

            found = False

            for i in range(len(self.orders[r])):

                if item == self.orders[r][i][0]:

                    self.orders[r].pop(i)

                    found = True

                    print(f"\n-------Removed {item} successfully------")

                    break

            if found == False:
                print("\n-----Item not found in the Bucket----")

            # remove restaurant if no items left
            if len(self.orders[r]) == 0:
                del self.orders[r]

        else:

            print("\nRestaurant not found in cart")

        print("Bucket =", self.orders)

    def cartvalue(self):

        self.count = 0

        for i in self.orders:

            for j in range(len(self.orders[i])):

                self.count += ((self.orders[i][j][1]) * (self.orders[i][j][2]))

        print(f"\n-------Total cart value={self.count}-------")

    def showcart(self):

        print("\n============== YOUR CART ==============")

        if len(self.orders) == 0:

            print("Cart is empty")

        else:

            for i in self.orders:

                print(f"\nRestaurant: {i}")

                for j in self.orders[i]:

                    print(
                        f"Item: {j[0]} | Price: {j[1]} | Quantity: {j[2]}"
                    )

    def bill(self):

        if len(self.orders) == 0:
            print("\nCart is empty")
            return

        self.cartvalue()

        print("\n##################################################################")

        print(f"Customer name:{self.cname}\nCustomer mobile number:{self.cphoneNo}")

        print("\n============== BILL ==============")

        for i in self.orders:

            print(f"\nRestaurant: {i}")

            for j in self.orders[i]:

                total = j[1] * j[2]

                print(
                    f"{j[0]} -> {j[1]} x {j[2]} = {total}"
                )

        print("----------------------------")
        print(f"Total Bill={self.count}")

        self.pay()

    def pay(self):

        print("\n\nDo you want to pay bill?:")

        x = input("enter Yes or No: ")

        if x.lower() == "yes":

            print("Payment successful!\nOrders placed successfully")
            self.orders={}

        else:

            print("Payment failed!\nOrder has been cancelled")


class entry:

    def __init__(self):

        self.obj = order()

    def start(self):

        while True:

            print("\n===================================")
            print("1. Add Item")
            print("2. Remove Item")
            print("3. Show Cart")
            print("4. Cart Value")
            print("5. Generate Bill")
            print("6. Exit")

            ch = input("\nEnter Choice: ")

            if ch == "1":

                self.obj.add()

            elif ch == "2":

                self.obj.remove()

            elif ch == "3":

                self.obj.showcart()

            elif ch == "4":

                self.obj.cartvalue()

            elif ch == "5":

                self.obj.bill()

            elif ch == "6":

                print("\nThank You!")
                break

            else:

                print("\nInvalid Choice!")


app = entry()
app.start()
```
